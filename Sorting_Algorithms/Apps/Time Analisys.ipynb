#### **1) Librarys**


```python
import random
import matplotlib.pyplot as plt
import time
```

#### **2) Algorithms**


```python
def InsertionSort(lista):
    for i in range(len(lista)):
        index = lista[i] 
        j = i-1 
        while(j >= 0 and lista[j] > index):
            lista[j + 1] = lista[j] 
            j -=1
        lista[j+1] = index
    return lista

def SelectionSort(list):
    for i in range(len(list)):
        min_idx = i
        
        for j in range(i+1,len(list)):
            if (list[j] < list[min_idx]):
                min_idx = j

        list[i], list[min_idx] = list[min_idx], list[i]
    return list

def BubbleSort_Optimized(list):
    for i in range(len(list),0,-1):  
        for j in range(i-1):
            if(list[j] > list[j+1]):
                list[j], list[j+1] = list[j+1], list[j]
    return list  
```

#### **3) Function for calculate the time**


```python
#Function who calculates de times for all lengths in n
def SortTime(func,num_samples):
    #Array with ejecution times inicialized with 0s
    timelist = []#[0 for i in range(len(num_samples))]
    for samples in num_samples:
        #Obtains the randon numbers
        list = [random.randint(-9999, 9999) for j in range(samples)]
        #Calculates the Sorting Time
        start = time.time()
        func(list)
        end = time.time()
        #add to timelist
        timelist.append(abs(start-end))
    return timelist
```

```python
#length of arrays
num_samples = [100,200,300,400,500,600,700,800,900,
               1000,2000,3000,4000,5000,6000,7000,8000,9000]
               #10000,11000,12000,12000,13000,14000,15000,16000,17000,18000,19000,
               #20000]
```

#### **3) Plotting**
 

```python
Time_Insert = SortTime(InsertionSort,num_samples);
```

```python
Time_Selection = SortTime(SelectionSort,num_samples);
```

```python
Time_Bubble = SortTime(BubbleSort_Optimized,num_samples);
```

```python
plt.figure(figsize=(20, 5))
plt.plot(num_samples, Time_Insert,   color='blue',  marker='o', markerfacecolor='pink', label='Insert')
plt.plot(num_samples,Time_Selection, color='green',  marker='o', markerfacecolor='aqua', label='Selection')
plt.plot(num_samples, Time_Bubble,   color='k',  marker='o', markerfacecolor='red', label='Bubble')
plt.ylabel('Time [s]')
plt.xlabel('Samples')
plt.title('Time Analysis')
plt.grid(True)
plt.legend()
plt.show()
```
